```python
# Phase 1: Strategic Imputation

data['CouponCode'] = data['CouponCode'].fillna('NO_COUPON')

print("Missing values remaining:", data['CouponCode'].isnull().sum())
```

```python
# Phase 2: Data Integrity - Cheecking For Duplicate Rows

duplicate_count=  data.duplicated().sum()
print(f" Found {duplicate_count} duplicate rows.")
```

```python
# Dropping Duplicates if any From Dataset

if duplicate_count > 0:
    data = data.drop_duplicates(keep='first')
    print("Dropped duplicate rows successfully.")
```

```python
# correct Data Formats

for col in ['Product', 'PaymentMethod', 'OrderStatus']:
    data[col] = data[col].astype(str).str.strip().str.title()
```

```python
# Convert the Date column into a uniform string layout (YYYY-MM-DD)

data['Date'] = pd.to_datetime(data['Date']).dt.strftime('%Y-%m-%d')
data.head(20)
```
