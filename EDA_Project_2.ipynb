```python
import pandas as pd
```

```python
# importing Dataset

data= pd.read_excel("C:/Users/priya/Downloads/Dataset for Data Analytics .xlsx")
data.head(10)
```

```python
 # Calculation of basic statistics (Count,Mean and Median)

columns= ['Quantity', 'UnitPrice', 'TotalPrice']
data[columns].describe().loc[['count', 'mean', '50%']]
```

```python
# Identify Trends and Outliers (Graphs)

import matplotlib.pyplot as plt
import seaborn as sns

# Box Plot for Outlier Detection in TotalPrice
plt.figure(figsize=(8, 4))
sns.boxplot(x=data['TotalPrice'], color='skyblue')
plt.title('Outlier Detection in Total Price')
plt.xlabel('Total Price')
plt.show()

# Histogram for Distribution of UnitPrice
plt.figure(figsize=(8, 4))
sns.histplot(data['UnitPrice'], kde=True, color='salmon')
plt.title('Distribution of Unit Prices')
plt.xlabel('Unit Price')
plt.show()

# Plot Bar Chart to visualize Sales Trend
product_sales = data.groupby('Product')['TotalPrice'].sum().reset_index()
plt.figure(figsize=(10, 5))
sns.barplot(data=product_sales, x='Product', y='TotalPrice', hue='Product', palette='viridis', legend=False)
plt.title('Total Revenue Generated by Each Product')
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```
